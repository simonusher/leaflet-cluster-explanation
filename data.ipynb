```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
from srai.regionalizers import geocode_to_region_gdf
```

```python
def get_points(region, size):
    boundary = geocode_to_region_gdf(region)
    return boundary["geometry"].sample_points(size=size).explode()
```

```python
my_map = folium.Map(location = (54.50272, 20.51721), zoom_start = 5)
my_map
```

```python
points = get_points("Poland", 350)
points.explore()
```

```python
pskov = geocode_to_region_gdf("Pskov")
pskov.explore()
```

```python
poland_points = get_points("Poland", 350)
pskov_points = get_points("Pskov", 25)
riga_points = get_points("Riga", 25)
tallin_points = get_points("Tallin", 25)
```

```python
marker_cluster = MarkerCluster().add_to(my_map)

points.apply(lambda row: folium.Marker(location=[row.y, row.x]).add_to(marker_cluster))
riga_points.apply(lambda row: folium.Marker(location=[row.y, row.x]).add_to(marker_cluster))
tallin_points.apply(lambda row: folium.Marker(location=[row.y, row.x]).add_to(marker_cluster))
pskov_points.apply(lambda row: folium.Marker(location=[row.y, row.x], icon=folium.Icon(color='red')).add_to(marker_cluster))

my_map
```
